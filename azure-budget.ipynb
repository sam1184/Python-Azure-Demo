```python
# build a cost calculator that mimics Azure Cost Management
resource_name = "prod-web-app"
resource_type = "App Service"
daily_cost = 45.50
monthly_budget = 1200.00

# Calculate projections
monthly_projection = daily_cost * 30
is_over_budget = monthly_projection > monthly_budget
budget_percentage = (monthly_projection / monthly_budget) * 100

print(f"Resource: {resource_name}")
print(f"Type: {resource_type}")
print(f"Monthly Projection: ${monthly_projection:.2f}")
print(f"Budget Status: {'⚠️ OVER BUDGET' if is_over_budget else '✅ Within Budget'}")
print(f"Budget Usage: {budget_percentage:.1f}%")

# Decision logic
if budget_percentage > 100:
    print("Action: Scale down or review resource")
elif budget_percentage > 80:
    print("Action: Monitor closely")
else:
    print("Action: No action needed")
```
